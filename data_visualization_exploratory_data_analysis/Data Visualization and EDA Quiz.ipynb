<h1>Data Visualization and EDA Quiz</h1>

<h3>Power Plant Cost</h3>
The "Nuclear" dataset contains information about the construction of several nuclear power plants across the US in the 1960s and 1970s. The dataset is located at <a href="https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/boot/nuclear.csv. The data description file is at https://stat.ethz.ch/R-manual/R-devel/library/boot/html/nuclear.html">Power Plant</a>. Read the data description to get some understanding of what's in the data. Load the dataset into pandas. You can either download it, or use the URL directly. The following questions are related to this dataset. What is the mean cost of a power plant, in millions of USD? Round your answer to 2 decimal points.

```python
power_plant_data = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/boot/nuclear.csv")
power_plant_data
```

```python
round(power_plant_data["cost"].mean(), 2)
```

<h3>Number of Stations</h3>
How many of all stations were constructed in the north-east region of the US?

```python
power_plant_data[power_plant_data["ne"] == 1].cost.count()
```

<h3>Cost over Time</h3>
What is the general behaviour of the cost of a power plant over time? Write "increasing", "decreasing" or "constant", without the quotes.

```python
# We can found out with scatterplot
# We can conclude that the cost increasing in time
plt.scatter(power_plant_data["date"], power_plant_data["cost"])
plt.xlabel("Time")
plt.ylabel("Cost [$ bln.]")
plt.show()
```

<h3>Max Power</h3>
Some stations are built on places where there were existing stations in the past. Out of all of these, how much power does the most powerful one generate, in MWe? Your answer should be a whole number.

```python
power_plant_data[power_plant_data["pr"] == 1].cap.max()
```
